```python
# Intersect Arrays and Fill Between Example

import matplotlib.pyplot as plt
import numpy as np

# Note: Arrays need to be the same size
x=np.arange(0,1,.1)
y1 = [1.0]*10
y2=np.arange(0,1,.1)
y3=[.6]*10

# Create and intersected Array
y4=np.maximum(y2,y3)


fig=plt.subplot()
plt.plot(x, y1, label='y1')
plt.plot(x, y2, label='y2')
plt.plot(x, y3, label='y3')
plt.fill_between(x, y1, y4)
plt.legend()
plt.title('Fill Between Arrays')
plt.show()

# Plot the intersected array
plt.plot(x, y4)
plt.title('Intersected Array')


```
